## In this data set we have to create a random test set as model needs to train on whole dataset

#### To create train and test data
1. Select groups based on Disease+Symptom_1+Symptom_2+Symptom_3
2. For each group pick some unique rows into training 
3. Keep Random Unique rows for testing

```python
import random
groups = symptoms[['Disease','Symptom_1','Symptom_2','Symptom_3']].drop_duplicates().copy()
train =pd.DataFrame(columns=symptoms.columns)
test=pd.DataFrame(columns=symptoms.columns)

groups.head(20)
```

```python
symptoms["ID"] = symptoms.index
symptoms.head()
```

```python
train_indexes_lst=[]
test_indexes_lst=[]
for i in groups.index:
    current_group=symptoms[(symptoms['Disease']==groups['Disease'][i]) & (symptoms['Symptom_1']==groups['Symptom_1'][i]) 
             & (symptoms['Symptom_2']==groups['Symptom_2'][i]) & (symptoms['Symptom_2']==groups['Symptom_2'][i])]
    train_length = math.floor(len(current_group['ID'])*97/100)
    set_indexes = set(current_group['ID'])
    train_indexes = set(random.sample(list(set_indexes),train_length))
    test_indexes = set_indexes.difference(train_indexes)
    for j in train_indexes:
        train_indexes_lst.append(j)
    for j in test_indexes:
        test_indexes_lst.append(j)

train_indexes_lst=list(set(train_indexes_lst))
test_indexes_lst=list(set(test_indexes_lst))
print(len(train_indexes_lst),len(test_indexes_lst))
```

##### Train and Test Dataset formation

```python
train_data = symptoms.iloc[train_indexes_lst,].drop("ID",axis=1)
test_data = symptoms.iloc[test_indexes_lst,].drop("ID",axis=1)
train_data.shape, test_data.shape
```

#### Encoding both train and test datasets

```python
train_data.head()
```

```python
train_data_new = pd.DataFrame(columns=lst)
for i in train_data.index:
    dct={}
    for k in lst:
        dct[k]=[0]
    dct['Disease'] = train_data['Disease'][i]
    for j in train_data.columns[1:]:
        if pd.isna(train_data[j][i]):
            continue
        else:
            dct[train_data[j][i].strip()]=[1]
    tmp=pd.DataFrame(dct)
    train_data_new = pd.concat([train_data_new,tmp],axis=0)
train_data_new.head()
```

```python
test_data_new = pd.DataFrame(columns=lst)
for i in test_data.index:
    dct={}
    for k in lst:
        dct[k]=[0]
    dct['Disease'] = test_data['Disease'][i]
    for j in test_data.columns[1:]:
        if pd.isna(test_data[j][i]):
            continue
        else:
            dct[test_data[j][i].strip()]=[1]
    tmp=pd.DataFrame(dct)
    test_data_new = pd.concat([test_data_new,tmp],axis=0)
test_data_new.head()
```

```python
#### Encoding target column
train_data_new['Target_disease'] = le.transform(train_data_new['Disease'])
test_data_new['Target_disease'] = le.transform(test_data_new['Disease'])

train_data_new.shape,test_data_new.shape
```
