```python
import pandas as pd

pd.options.display.max_rows = None
pd.options.display.max_columns = None

import warnings 
warnings.filterwarnings("ignore")
import os
print(os.getcwd())
```

```python
PROCESSED_DATA_DIR=os.getcwd()+"\\data\\processed"
train_data = pd.read_csv(PROCESSED_DATA_DIR+"\\train_dataset.csv")
test_data = pd.read_csv(PROCESSED_DATA_DIR+"\\test_dataset.csv")
train_data.shape,test_data.shape
```

```python
train_data.columns
```

```python
X_train= train_data.drop(['Disease','Target_disease'],axis=1)
y_train= train_data['Target_disease']

X_test = test_data.drop(['Disease','Target_disease'],axis=1)
y_test = test_data['Target_disease']
```

#### Creating metrics for  Classification model

```python
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score, precision_score
def evaluation(y_true,y_pred):
    print("Confusion Matrix:")
    print(confusion_matrix(y_true,y_pred))
    print("\n Accuracy score:")
    print(accuracy_score(y_true,y_pred))
    print("\n Precision:")
    print(precision_score(y_true,y_pred,average=None))
```

#### 1. Creating a Decision tree model 

```python
from sklearn.tree import DecisionTreeClassifier

dtree = DecisionTreeClassifier(random_state=100)
dtree_fit = dtree.fit(X_train,y_train)
```

```python
y_train_pred = dtree.predict(X_train)
evaluation(y_train,y_train_pred)
```

```python
y_test_pred = dtree.predict(X_test)
evaluation(y_test,y_test_pred)
```

##### 2. Decision Tree is correctly predicting all symptoms 

```python
## Reading the total given dataset
symptoms = pd.read_csv(PROCESSED_DATA_DIR+"\\All_symptoms.csv")
symptoms.head()
```

```python
X= symptoms.drop(['Disease','Encoded_disease'],axis=1)
y= symptoms['Encoded_disease']
```

```python
#### 2. Let us train on the whole dataset
dtree_all = DecisionTreeClassifier()
dtree_all_fit = dtree_all.fit(X,y)
```

```python
## Evaluate
y_pred = dtree_all.predict(X)
evaluation(y,y_pred)
```

```python
# Saving the model 
import pickle
MODELS_DIR = os.getcwd()+"\\models"
model_file = open(MODELS_DIR+"\\dtree_model.save" ,"wb")
pickle.dump(dtree_all,model_file)
model_file.close()
```
